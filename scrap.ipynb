**Getting the html page for each year**

```python
def get_team_stat(years=years):
    for year in years : 
        print(f"Extracting data from year : {year} ...")
        url_year = url.format(year)

        driver.get(url_year)
        driver.execute_script("window.scrollTo(1,10000)")
        time.sleep(2)
        html_page = driver.page_source

        with open(f'team_html/{year}.html', 'w+') as f:
            f.write(html_page)

        print(f'Year : {year} Done!\n')
        print('-------------------------------')
```

```python
import re
```

```python
def get_team_stat(filename):
    with open(filename, 'r') as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')

    length = len(soup.find_all(name='tr', attrs={'class':'thead'}))

    for i in range(length):
        soup.find(name='tr', attrs={'class':'thead'}).decompose()

    team_table = soup.find(id='divs_standings_E')
    team_df_E = pd.read_html(str(team_table))[0]
    team_df_E['year'] = re.findall('[0-9]+', filename)[0]
    team_df_E['conference'] = 'Eastern'
    team_df_E.rename(columns={'Eastern Conference':'team'}, inplace=True)
    
    team_table = soup.find(id='divs_standings_W')
    team_df_W = pd.read_html(str(team_table))[0]
    team_df_W['year'] = re.findall('[0-9]+', filename)[0]
    team_df_W['conference'] = 'Western'
    team_df_W.rename(columns={'Western Conference':'team'}, inplace=True)
    
    team_df = pd.concat([team_df_E, team_df_W])
    team_df['team'] = team_df['team'].apply(lambda x: str(x).replace('*', ''))
    team_df.rename(columns={'W' : 'wins'}, inplace=True)
    
    expanded_table = soup.find(name='table', attrs={'id' : 'expanded_standings'})
    expanded_df = pd.read_html(str(expanded_table))[0]
    expanded_df.columns = expanded_df.columns.droplevel()
    expanded_df.rename(columns={'Team':'team'}, inplace=True)
    
    cols_to_transform = list(expanded_df.columns)[list(expanded_df.columns).index('≥10'):]
    #print(cols_to_transform)
    dict_ = { cols: f'month {i}' 
         for cols, i in zip(cols_to_transform, range(len(cols_to_transform)))
         if cols != '≥10'
        }
    expanded_df.rename(columns=dict_, inplace=True)
    #print(dict_)
    
    final_df = team_df.merge(expanded_df, on=['team'])
    #return final_df
    return final_df, team_df , expanded_df
```
